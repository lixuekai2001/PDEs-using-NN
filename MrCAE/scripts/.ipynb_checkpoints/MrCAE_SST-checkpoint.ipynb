```python
# imports
import os
import sys
import torch
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib notebook

module_path = os.path.abspath(os.path.join('../src/'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import torch_cae_multilevel_V4 as net
```

### MrCAE

```python
# init model & load data
data_path = '../data/npy/sst.mnmean.mask2.npy'
model_path = '../model/sst/new'
result_path = '../result/sst/'

dataset = net.MultiScaleDynamicsDataSet(data_path, n_levels=5)
model = net.CAE(n_levels=5, activation=torch.nn.Sequential())
```

```python
# training
# -- level 0 --
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=5000, batch_size=315, widen_sizes=[2,2,2], 
                                                             tol=0.01, w=0.5, verbose=2, std=0.01, err_decrease_thresh=1e-6, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l0.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l0.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l0.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```

```python
# -- level 1 --
# model = torch.load(os.path.join(model_path, 'model_L0_2.pt'))
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=5000, batch_size=315, widen_sizes=[5,5,5,5], 
                                                             tol=0.08, w=0.5, verbose=2, std=0.01, err_decrease_thresh=1e-6, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l1.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l1.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l1.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```

```python
# -- level 2 --
# model = torch.load(os.path.join(model_path, 'model_L1_2.pt'))
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=5000, batch_size=315, widen_sizes=[10,10,10,10], 
                                                             tol=0.05, w=0.5, verbose=2, std=0.01, err_decrease_thresh=1e-6, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l2.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l2.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l2.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```

```python
# -- level 3 --
# model = torch.load(os.path.join(model_path, 'model_L2_3.pt'))
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=5000, batch_size=315, widen_sizes=[10,10,10,10,10], 
                                                             tol=0.02, w=0.5, verbose=2, err_decrease_thresh=1e-6, std=0.01, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l3.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l3.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l3.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```

```python
# -- level 4 --
# model = torch.load(os.path.join(model_path, 'model_L3_3.pt'))
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=5000, batch_size=315, widen_sizes=[10,10,10,10,20,20], 
                                                             tol=0.01, w=0.5, verbose=2, err_decrease_thresh=1e-6, std=0.01, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l4.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l4.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l4.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```

### Results

```python
# trained models at different levels
models = {}
print('model names: model_L{level}_{index}')
for file_name in sorted(os.listdir(model_path)):
    model_name, _ = file_name.split('.')
    print(model_name)
    models[model_name] = torch.load(os.path.join(model_path, file_name))
```

```python
# choose the finest model
model = models['model_L4_2']

# resolved maps at different levels (that suggest poorly reconstructed regions)
for i in range(5):
    print(model.resolved_maps[str(i)].keys())
    
for i in range(5):
    level_maps = model.resolved_maps[str(i)]
    for key in level_maps.keys():
        if key == '0':
            intensities = (1 - level_maps[key])
        else:
            intensities += (1 - level_maps[key])
    file_name = 'L{}.png'.format(i)
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    ax.imshow(intensities.cpu().detach().numpy(), cmap='hot')
    ax.set_xticks([])
    ax.set_yticks([])
    fig.savefig(os.path.join(result_path, file_name))
```

```python
inds = np.array(sorted(dataset.test_inds))
dataset.test_inds = inds
n_snapshots = len(inds)
n_samples = 4
n_step = n_snapshots // 4

fig, axes = plt.subplots(1, n_samples, figsize=(n_samples*14, 10))
plt.subplots_adjust(wspace=0.01)
for i in range(n_samples):
    axes[i].imshow(dataset.data[inds[i*n_step], :, :, :].squeeze().cpu().detach().numpy(), cmap='viridis')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
fig.savefig(os.path.join(result_path, 'test_samples.png'), bbox_inches='tight')


model = models['model_L0_1']
fig, axes = plt.subplots(1, n_samples, figsize=(n_samples*14, 10))
plt.subplots_adjust(wspace=0.01)
for i in range(n_samples):
    _, _, data = dataset.obtain_data_at_current_level(level=0)
    output, _, _, _ = model(data[[i*n_step], :, :, :], model.cur_level)
    axes[i].imshow(output.squeeze().cpu().detach().numpy(), cmap='viridis')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    
fig.savefig(os.path.join(result_path, 'test_L0_reconstructions.png'), bbox_inches='tight')


model = models['model_L1_1']
fig, axes = plt.subplots(1, n_samples, figsize=(n_samples*14, 10))
plt.subplots_adjust(wspace=0.01)
for i in range(n_samples):
    _, _, data = dataset.obtain_data_at_current_level(level=1)
    output, _, _, _ = model(data[[i*n_step], :, :, :], model.cur_level)
    axes[i].imshow(output.squeeze().cpu().detach().numpy(), cmap='viridis')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
fig.savefig(os.path.join(result_path, 'test_L1_reconstructions.png'), bbox_inches='tight')


model = models['model_L2_1']
fig, axes = plt.subplots(1, n_samples, figsize=(n_samples*14, 10))
plt.subplots_adjust(wspace=0.01)
for i in range(n_samples):
    _, _, data = dataset.obtain_data_at_current_level(level=2)
    output, _, _, _ = model(data[[i*n_step], :, :, :], model.cur_level)
    axes[i].imshow(output.squeeze().cpu().detach().numpy(), cmap='viridis')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
fig.savefig(os.path.join(result_path, 'test_L2_reconstructions.png'), bbox_inches='tight')

model = models['model_L3_2']
fig, axes = plt.subplots(1, n_samples, figsize=(n_samples*14, 10))
plt.subplots_adjust(wspace=0.01)
for i in range(n_samples):
    _, _, data = dataset.obtain_data_at_current_level(level=3)
    output, _, _, _ = model(data[[i*n_step], :, :, :], model.cur_level)
    axes[i].imshow(output.squeeze().cpu().detach().numpy(), cmap='viridis')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
fig.savefig(os.path.join(result_path, 'test_L3_reconstructions.png'), bbox_inches='tight')

model = models['model_L4_2']
fig, axes = plt.subplots(1, n_samples, figsize=(n_samples*14, 10))
plt.subplots_adjust(wspace=0.01)
for i in range(n_samples):
    _, _, data = dataset.obtain_data_at_current_level(level=4)
    output, _, _, _ = model(data[[i*n_step], :, :, :], model.cur_level)
    axes[i].imshow(output.squeeze().cpu().detach().numpy(), cmap='viridis')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
fig.savefig(os.path.join(result_path, 'test_L4_reconstructions.png'), bbox_inches='tight')
```

```python
records = {}
for file_name in sorted(os.listdir(result_path)):
    if file_name.endswith('.dat'):
        key, _ = file_name.split('.')
        with open(os.path.join(result_path, file_name), 'rb') as f: 
            records[key]= pickle.load(f)
```

```python
n_colors = 4
colors = [(41/255, 175/255, 29/255)]+ \
         [(i/255, j/255, k/255) for i, j, k in 
          zip(np.linspace(47, 17, n_colors), 
              np.linspace(230, 123, n_colors), 
              np.linspace(207, 118, n_colors))]

fig1, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(80, 16))
fig2, ax2 = plt.subplots(nrows=1, ncols=1, figsize=(80, 16))

# concantenate time
all_times = []
for i in range(5):
    level_times = records['times_l{}'.format(i)]
    for j in range(len(level_times)):
        all_times += level_times[j]
all_times = list(np.array(all_times).cumsum())

# plot
s = 0
for i in range(5):
    level_errs = records['val_errs_l{}'.format(i)]
    n_widens = len(level_errs)
    ax1.axvline(x=s, color='k', linestyle='--', linewidth=20)
    ax2.axvline(x=all_times[s], color='k', linestyle='--', linewidth=20)
    for j in range(n_widens):
        op_err = level_errs[j]
        ax1.plot(range(s, s + len(op_err)), np.log(op_err), color=colors[j], linewidth=20)
        ax2.plot(all_times[s:s + len(op_err)], np.log(op_err), color=colors[j], linewidth=20)
        s += len(op_err)
        
ax1.axvline(x=s-1, color='k', linestyle='--', linewidth=20)
ax2.axvline(x=all_times[s-1], color='k', linestyle='--', linewidth=20)

ax1.xaxis.set_tick_params(labelsize=100)
ax1.yaxis.set_tick_params(labelsize=100)
ax2.xaxis.set_tick_params(labelsize=100)
ax2.yaxis.set_tick_params(labelsize=100)
ax1.set_ylim([-5, 0])
ax2.set_ylim([-5, 0])

fig1.savefig(os.path.join(result_path, 'err_iter_plot.png'))
fig2.savefig(os.path.join(result_path, 'err_time_plot.png'))
```
