```python
# imports
import os
import cv2
import sys
import torch
import pickle
# import pyJHTDB
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

%matplotlib notebook

module_path = os.path.abspath(os.path.join('../src/'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import torch_cae_multilevel_V4 as net
```

```python
if os.path.exists('../data/npy/channel_flow.npy'):
    data = np.load('../data/npy/channel_flow.npy')
else:
    # load shared library
    lTDB = pyJHTDB.libJHTDB()
    #initialize webservices
    lTDB.initialize()

    # add token
    auth_token  = "edu.uw.yliu814-dff25860"  # get ur own token from: http://turbulence.pha.jhu.edu/authtoken.aspx
    lTDB.add_token(auth_token)
    
    # info
    n = 3
    info = pyJHTDB.dbinfo.channel
    name = info['name']
    x_sample_factor = 4
    y_sample_factor = 4
    nx = info['nx'] // x_sample_factor
    ny = info['ny'] // y_sample_factor
    xnodes = info['xnodes'][::x_sample_factor]
    ynodes = info['ynodes'][::y_sample_factor]
    zmid = 1.5*np.pi
    
    lpoints = []
    for i in range(nx):
        for j in range(ny):
            lpoints.append([xnodes[i], ynodes[j], zmid])
            
    points = np.array(lpoints, dtype='float32')
    
    data = []
    for t in info['time'][::8]:
        sample = lTDB.getData(t, points, data_set = name, getFunction = 'getVelocity')
        sample = np.reshape(sample, (nx, ny, n))
        print('time t: ', t, ', sample shape: ', sample.shape)
        data.append(sample)
    
    data = np.stack(data, 0)
    Phi = np.squeeze(data[:, :-1, :-1, 0])
    data = (Phi - Phi.min()) / (Phi.max() - Phi.min())
    np.save('../data/npy/channel_flow.npy', data)
    
# viz
plt.figure(figsize=(5, 10))
plt.imshow(data[1, :, :])
plt.title('sampled snapshot',fontsize=20)
```

```python
# init model & load data
data_path = '../data/npy/channel_flow.npy'
model_path = '../model/fluid2/new'
result_path = '../result/fluid2/'

dataset = net.MultiScaleDynamicsDataSet(data_path, n_levels=4)
model = net.CAE(n_levels=4, activation=torch.nn.Sequential())
```

```python
# training
# -- level 0 --
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=3000, batch_size=350, widen_sizes=[4,4,4,4], 
                                                             tol=0.02, w=0.5, verbose=2, std=0.01, err_decrease_thresh=1e-6, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l0.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l0.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l0.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```

```python
# -- level 1 --
model = torch.load(os.path.join(model_path, 'model_L0_2.pt'))
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=3000, batch_size=350, widen_sizes=[8,8,8,8,8,8,8], 
                                                             tol=0.01, w=0.5, verbose=2, std=0.01, err_decrease_thresh=1e-6, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l1.dat'), "wb") as fp:
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l1.dat'), "wb") as fp:
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l1.dat'), "wb") as fp:
    pickle.dump(ts, fp)
```

```python
# -- level 2 --
# model = torch.load(os.path.join(model_path, 'model_L0_2.pt'))
train_errs, val_errs, ts = model.train_on_this_level_wrapper(dataset, max_epoch=3000, batch_size=350, widen_sizes=[16,16,16,16,16,16,16,16], 
                                                             tol=0.008, w=0.5, verbose=2, std=0.01, err_decrease_thresh=1e-6, 
                                                             model_path=model_path)

# dump the results
with open(os.path.join(result_path, 'train_errs_l2.dat'), "wb") as fp: 
    pickle.dump(train_errs, fp)
with open(os.path.join(result_path, 'val_errs_l2.dat'), "wb") as fp: 
    pickle.dump(val_errs, fp)
with open(os.path.join(result_path, 'times_l2.dat'), "wb") as fp: 
    pickle.dump(ts, fp)
```
