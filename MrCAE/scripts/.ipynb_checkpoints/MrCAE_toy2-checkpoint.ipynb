## Toy Model 2: traveling wave

### Created by Yuying Liu, 09/23/2019

$$\Phi(x, t) = u(x)cos(\omega_0 t) + v(x, t)cos(\omega_1 t + \frac{\pi}{4})$$

```python
# imports
import os
import sys
import torch
import numpy as np
import matplotlib.pyplot as plt

module_path = os.path.abspath(os.path.join('../src/'))
if module_path not in sys.path:
    sys.path.append(module_path)
    
import torch_cae_multilevel_V3 as net
```

## simulation

```python
# params
w0 = 0.5
w1 = 4.0
sigma0 = 10.0
sigma1 = 0.25
v = 0.5
```

```python
# define two modes
def phi1(x, y, t):
    return 1./np.cosh((x+1)/sigma0)/np.cosh((y-1)/sigma0)*np.cos(w0*t)

def phi2(x, y, t):
    return 1./(sigma1*np.sqrt(2*np.pi))*np.exp(-((x-3+v*t)**2+(y+3-v*t)**2)/(2*sigma1**2))*np.cos(w1*t)
```

```python
# mesh grids
x = np.linspace(-5, 5, 127)
y = np.linspace(-5, 5, 127)
t = np.linspace(0, 4*np.pi, 500)
xgrid, ygrid, tgrid = np.meshgrid(x, y, t)
```

```python
# viz modes
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].pcolor(xgrid[:,:,0], ygrid[:,:,0], phi1(xgrid, ygrid, tgrid)[:, :, 0])
axes[1, 0].pcolor(xgrid[:,:,0], ygrid[:,:,0], phi2(xgrid, ygrid, tgrid)[:, :, 0])
axes[0, 1].plot(t, np.cos(w0*t))
axes[1, 1].plot(t, np.cos(w1*t + np.pi/4))
#
axes[0, 0].set_ylabel('mode 1', fontsize=16)
axes[1, 0].set_ylabel('mode 2', fontsize=16)
axes[0, 0].set_title('spatial modes', fontsize=16)
axes[0, 1].set_title('temporal modes', fontsize=16)
```

```python
# data
Phi = phi1(xgrid, ygrid, tgrid) + phi2(xgrid, ygrid, tgrid)
scaled_Phi = (Phi - Phi.min()) / (Phi.max() - Phi.min())
print(Phi.shape)
```

```python
# save 
np.save('../data/npy/toy2.npy', scaled_Phi.T)
```

```python
# viz overall dynamics
step = 500 // 16
fig = plt.figure(figsize=(18, 12))
for i in range(16):
    plt.subplot(4, 4, i+1)
    plt.pcolor(xgrid[:,:,0], ygrid[:,:,0], Phi[:,:,i*step])
```

## MrCAE

```python
# init model & load data
data_path = '../data/npy/toy2.npy'
model_path = '../model/toy2/'
result_path = '../result/toy2/'

scaled_Phi = np.load(data_path)
model = net.CAE(n_levels=3, n_blocks=3, activation=torch.nn.ReLU())
model.load_data(scaled_Phi)
```

```python
# training
# -- level 0 --
_, _, _ = model.train_on_this_level_wrapper(max_epoch=10000, batch_size=350, widen_sizes=[4,4,4], tol=0.001, 
                                            w=0.5, verbose=2, std=0.01, err_decrease_thresh=None, model_path=model_path)
```

```python
# -- level 1 --
_, _, _ = model.train_on_this_level_wrapper(max_epoch=10000, batch_size=350, widen_sizes=[4,4,4,4], tol=0.0008, 
                                            w=0.5, verbose=2, std=0.01, err_decrease_thresh=None, model_path=model_path)
```

```python
# -- level 2 --
_, _, _ = model.train_on_this_level_wrapper(max_epoch=10000, batch_size=350, widen_sizes=[4,4,4,4,4,4], tol=0.0006, 
                                            w=0.5, verbose=2, std=0.01, err_decrease_thresh=None, model_path=model_path)
```

## Results

### Linear CAE

```python
# trained models at different levels
models = {}
print('model names: model_L{level}_{index}')
for file_name in sorted(os.listdir('../model/toy2/linear/')):
    model_name, _ = file_name.split('.')
    print(model_name)
    models[model_name] = torch.load(os.path.join('../model/toy2/linear/', file_name))
```

```python
# choose the finest model
model = models['model_L2_0']

# resolved maps at different levels (that suggest poorly reconstructed regions)
for i in range(3):
    print(model.resolved_maps[str(i)].keys())

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].pcolor(model.test[2, 0, :, :].squeeze().cpu().detach().numpy())
axes[1].pcolor(model.resolved_maps['0']['0'].cpu().detach().numpy(), vmin=0, vmax=1)
```

```python
# choose a test snapshot
test_idx = 3
test_reconstructions = {}
test_snapshots = {}
for model_name in models.keys():
    model = models[model_name]
    _, _, data = model.obtain_data_at_current_level()
    output, _, _, _ = model(data[[test_idx], :, :, :], model.cur_level)
    test_reconstructions[model_name] = output.squeeze().cpu().detach().numpy()
    test_snapshots[model_name] = data[[test_idx], :, :, :].squeeze().cpu().detach().numpy()

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].pcolor(test_snapshots['model_L0_1'])
axes[1].pcolor(test_snapshots['model_L1_1'])
axes[2].pcolor(test_snapshots['model_L2_0'])

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].pcolor(test_reconstructions['model_L0_1'])
axes[1].pcolor(test_reconstructions['model_L1_1'])
axes[2].pcolor(test_reconstructions['model_L2_0'])
```

```python
op = torch.nn.Hardshrink(lambd=0.05)
_, _, _, hids = model.forward(model.ori_data, model.cur_level, query_hidden=True)
print('convolved feature keys: ', hids.keys())

step = 500 // 16
fig = plt.figure(figsize=(18, 12))
for i in range(16):
    plt.subplot(4, 4, i+1)
    activation_area = (op(hids['L1_1'][i*step, 0, :, :]) != 0).float()
    plt.pcolor(activation_area.squeeze().cpu().detach().numpy(), vmin=0, cmap='hot')
```
