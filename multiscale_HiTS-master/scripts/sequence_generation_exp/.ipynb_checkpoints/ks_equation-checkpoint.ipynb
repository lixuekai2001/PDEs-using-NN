## Part I

### process data

```python
obj = loadmat(os.path.join(data_dir, 'kuramoto_sivishinky.mat'))
data = obj['data']
print(data.shape)
np.save(os.path.join(data_dir, 'data.npy'), data)
```

```python
plt.figure(figsize=(16, 3))
plt.axis('off')
plt.imshow(data)
```

### coupled nn

```python
coupled_nn_pred = np.load(os.path.join(data_dir, 'KS_couple_pred.npy'))
print(((coupled_nn_pred.T - data)**2).mean())
plt.figure(figsize=(16, 3))
plt.axis('off')
plt.imshow(coupled_nn_pred.T)
```
