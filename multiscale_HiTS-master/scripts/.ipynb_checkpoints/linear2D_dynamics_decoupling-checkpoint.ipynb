### simulation

```python
# # # # # # # # # # # # # # # # # # # # 
#  global constants, parameters, etc. #
# # # # # # # # # # # # # # # # # # # #

np.random.seed(1)
dt = 0.001;
nsteps = 2**14;

n = 4
A = np.array([[0, 8, 50, 0], [-8, 0, 0, 0], [-50, 0, 0, 8], [0, 0, -8, 0]])

# # # # # # # # # # # # # # # # # 
# simulate harmonic oscillator  #
# # # # # # # # # # # # # # # # # 

def harmonic_oscillator_rhs(x):
    # rhs of harmonic oscillator
    return A.dot(x)

# simulate a trial (long)
print('generating a trial ...')
x_init = np.random.uniform(-1.0, 1.0, n)
t = np.linspace(0, (nsteps-1)*dt, nsteps)
sol = sp.integrate.solve_ivp(lambda _, x: harmonic_oscillator_rhs(x), [0, (nsteps-1)*dt], x_init, t_eval=t)
data = sol.y
print(data.shape)

# # # # # # # # # 
# visualization #
# # # # # # # # #

# visualize trajectories (in phase space)
print('visualizing ...')
fig = plt.figure(figsize=(16, 4))
gs = gridspec.GridSpec(nrows=1, ncols=1)
ax1 = fig.add_subplot(gs[0, 0])
for i in range(n):
    ax1.plot(data[i,:], linewidth=2.0)
ax1.set_title('Sampled Trajectories', fontsize=20)
```

### multiscale timestepper

```python
n_scales = 2
factor = 32
model = net.MultiScaleTimeStepper(n_scales=n_scales, factor=factor, arch=[n, 10, 20, 40, 20, 10, n], activation=torch.nn.ReLU())
model.load_data(data, dt)
```

```python
model.train_net(max_epoch=10000, batch_size=5000, lr=0.0005)
```

```python
nt = factor**0
x0 = model.train[:, -nt:]
forecast = model.forecast(x0, n_steps=2**10)
fig = plt.figure(figsize=(16, 4))
plt.plot(forecast.transpose(0, 1).detach().numpy())
```

```python
nt = factor**1
x0 = model.train[:, -nt:]
forecast = model.forecast(x0, n_steps=2**8)
fig = plt.figure(figsize=(16, 4))
plt.plot(forecast.transpose(0, 1).detach().numpy())
```
