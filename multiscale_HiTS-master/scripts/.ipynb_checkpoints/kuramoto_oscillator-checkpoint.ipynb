### simulation

```python
# # # # # # # # # # # # # # # # # # # # 
#  global constants, parameters, etc. #
# # # # # # # # # # # # # # # # # # # #

np.random.seed(2)
dt = 0.001
nsteps = 100
nsteps = 100
test_nsteps = 1000

K = 2
N = 9
omegas = np.random.uniform(0,1,N)

# # # # # # # # # # # # # # # # # 
#  simulate kuramoto oscillator #
# # # # # # # # # # # # # # # # #

def kuramoto_oscillator(theta, omegas, K, N):
    # rhs of kuramoto oscillator
    dtheta = omegas
    for i in range(N):
        for j in range(N):
            dtheta[i] = dtheta[i] + K/N*np.sin(theta[j]-theta[i])
    return dtheta


# simulate trials with uniscale sampling
print('generating trials in uniscale ...')
uni_states = np.zeros([uni_nsteps,uni_numICs,N])
for i in range(uni_numICs):
    theta_init = np.random.uniform(-2,2,N)
    t = np.linspace(0, (uni_nsteps-1)*uni_dt, uni_nsteps)
    sol = sp.integrate.solve_ivp(lambda _, theta: kuramoto_oscillator(theta, omegas, K, N), [0, (uni_nsteps-1)*uni_dt], theta_init, t_eval=t)
    uni_states[:,i,:] = sol.y.T

# simulate trials with multiscale sampling
print('generating trials in multiscales ...')
multi_states = np.zeros([multi_nsteps,multi_numICs,N])
dts = []
for i in range(multi_numICs):
    theta_init = np.random.uniform(-2,2,N)
    multi_dt = np.random.uniform(0.01,0.1)
    dts.append(multi_dt)
    t = np.linspace(0, (multi_nsteps-1)*multi_dt, multi_nsteps)
    sol = sp.integrate.solve_ivp(lambda _, theta: kuramoto_oscillator(theta, omegas, K, N), [0, (multi_nsteps-1)*multi_dt], theta_init, t_eval=t)
    multi_states[:,i,:] = sol.y.T

# simulate test trials (long)
print('generating test trials ...')
test_states = np.zeros([test_nsteps,test_numICs,N])
for i in range(test_numICs):
    theta_init = np.random.uniform(-2,2,N)
    t = np.linspace(0, (test_nsteps-1)*uni_dt, test_nsteps)
    sol = sp.integrate.solve_ivp(lambda _, theta: kuramoto_oscillator(theta, omegas, K, N), [0, (test_nsteps-1)*uni_dt], theta_init, t_eval=t)
    test_states[:,i,:] = sol.y.T


# # # # # # # # # # # # # 
#  collect data & save  #
# # # # # # # # # # # # # 

print('collecting and saving data ...')
my_dict = {}
my_dict['system'] = {'name': 'Kuramoto Oscillator', 'parameters': {'K': K, 'N': N, 'omegas': omegas.tolist()}}
my_dict['uni_scale'] = {'data': uni_states.tolist(), 'dt': uni_dt, 'nsteps': uni_nsteps}
my_dict['multi_scale'] = {'data': multi_states.tolist(), 'dts': dts, 'nsteps': multi_nsteps}
my_dict['test_trials'] = {'data': test_states.tolist(), 'dt': uni_dt, 'nsteps': test_nsteps}
with open('./data/Kuramoto/my_dict.json', 'w') as fp:
    json.dump(my_dict, fp)

# # # # # # # # # 
# visualization #
# # # # # # # # #

print('visualizing ...')

theta = test_states[:,0,:].T

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='polar')
color = np.random.rand(9,3) 
pts = []
for i in range(9):
    pt, = ax.plot(theta[i,0], 1.0, 'o', c=color[i,:], markersize=10)
    pts.append(pt)

def kura_update(i, pts):
    for k in range(9):
        pts[k].set_data(theta[k,i], 1.0)
    return (pts,)
    
anim = animation.FuncAnimation(fig, kura_update, frames=range(test_nsteps), interval=150, fargs=(pts,))
HTML(anim.to_html5_video())
```
